```python
number_of_images = {}

for dir in os.listdir(train_dir):
    number_of_images[dir] = len(os.listdir(os.path.join(train_dir, dir)))
    print(dir , " ",number_of_images[dir]) 
```

```python
def preprocessingImage1(path):
  image_data = ImageDataGenerator(zoom_range=0.2,shear_range=0.2,preprocessing_function= preprocess_input,horizontal_flip=True)
  image = image_data.flow_from_directory(directory=path,target_size=(224,224),batch_size=32,class_mode='binary')
  return image
```

```python
def preprocessionfImage2(path):
  image_data  = ImageDataGenerator(preprocessing_function= preprocess_input )
  image = image_data.flow_from_directory(directory=path,target_size=(224,224),batch_size=32,class_mode='binary')
  return image
```

```python
def datafolder(path,split):
  if not os.path.exists("./Data/"+path):
    os.mkdir("./Data/"+path)

    for dir in os.listdir(train_dir):
      os.makedirs("./Data/"+path+"/"+dir)
      for img in np.random.choice(a=os.listdir(os.path.join(train_dir,dir)),
                                  size=(math.floor(split * number_of_images[dir])-5),replace=False):
      
          O = os.path.join(train_dir,dir,img)
          D = os.path.join("./Data/"+path,dir)
          shutil.copy(O,D)
          os.remove(O)

  else:
     print("Folder already exist")
```

```python
datafolder("val",0.2)
```

```python
train_data = preprocessingImage1(train_dir)
```

```python
val_dir = 'Data/val/'
val_data = preprocessionfImage2(val_dir)
```

```python
test_dir = 'Data/test/'
test_data = preprocessionfImage2(test_dir)
```

```python
base_model = MobileNet(input_shape=(224,224,3),include_top=False)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
x= Flatten()(base_model.output)
x= Dense(units=1,activation='sigmoid')(x)
```

```python
model = Model(base_model.input,x)
```

```python
model.compile(optimizer='rmsprop', loss=keras.losses.binary_crossentropy, metrics=['accuracy'])
```

```python
mc = ModelCheckpoint(filepath="model.keras",monitor='val_accuracy',verbose=1,save_best_only=True)

#Early check points
es = EarlyStopping(monitor="val_accuracy",min_delta=0.01,patience=5,verbose=1)

cb = [mc,es]
```

```python
hist = model.fit(train_data,
                 steps_per_epoch=10,
                 epochs=30,
                 validation_data=val_data,
                 validation_steps=16,
                 callbacks=cb)
```

```python
model = load_model("model.keras")
```

```python
acc = model.evaluate(test_data)[1]
```

```python
print(f"model accuracy is {acc *100} %")
```
